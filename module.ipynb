#### Lesson 4　Chapter 3.2　標準ライブラリの活用

+ math モジュール

平方根を計算する: math.sqrt()

指数を計算する: math.exp()（自然対数の底 e の指数関数）

定数: math.pi（円周率）、math.e（自然対数の底）

```python
import math

# 平方根の計算
print(math.sqrt(25)) 

# 指数関数の計算
print(math.exp(1))  # 2.718281828459045（自然対数の底 e）

# 定数の使用
print(math.pi)  # 3.141592653589793
```

+ datetime モジュール

```python
import datetime as dt

# 現在の日付と時刻を取得
now = dt.datetime.now()
print(now)

# 特定の日付を作成
birthday = dt.datetime(1995,10,10)
print(birthday)


#日付の差を計算
date1 = dt.datetime(2000,5,10)
date2 = dt.datetime.now()
diff = date2 - date1
print(diff)
print(diff.days)
```

+ random モジュール

```python
import random

# ランダムな整数を生成
print(random.randint(1,10)) #1~10 ※1も10も含む　からランダム整数

# リストをシャッフル
my_list = [1, 2, 3, 4, 5]
random.shuffle(my_list)
print(my_list)  # 順番がランダムで出力

# リストからランダムに要素を選択
print(random.choice(my_list))  # リストからランダムに1つ選ばれる
```

+ urllib モジュール

```python
"""
Webリソースにアクセスするための機能を提供します。
特に、Webページのデータを取得するために利用され、スクレイピングや
APIの呼び出しに役立ちます。
"""
import urllib.request

# URLを指定してデータを取得
url = "https://techacademy.jp"
response = urllib.request.urlopen(url)
print(response,"\n")

# HTMLデータを読み込み、UTF-8文字列にデコード
html = response.read().decode("utf-8")
print(html[:300])  # 取得したHTMLの先頭300文字を表示
```

自作モジュールをインポートして使う
作成したモジュールを使いたいときは、import 文を使用します。

```python
import math_operations #a+b,a-bを返すモジュール

# モジュール内の関数を使う
result1 = math_operations.add(5, 3)
result2 = math_operations.subtract(10, 4)

print("5 + 3 =", result1)       # 出力: 5 + 3 = 8
print("10 - 4 =", result2)      # 出力: 10 - 4 = 6
```

```python
"""
直接実行された場合と、他のプログラムからインポートされた場合で
動作を変えるために、if __name__ == "__main__": という構文を使う。
モジュールのテストやデバッグの際に便利。
"""

def add(a, b):
    """2つの数を加算する"""
    return a + b

def subtract(a, b):
    """2つの数を減算する"""
    return a - b

# テスト用のコード
if __name__ == "__main__":
    # モジュールが直接実行された場合にだけ実行される
    print("5 + 3 =", add(5, 3))
    print("10 - 4 =", subtract(10, 4))
```
