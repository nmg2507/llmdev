#### Lesson 3　Chapter 5.1　タプルを作成する

複数の値をひとまとめにして扱えるデータ型。変更ができない。

```python
char = ("a", "b", "c")
numbers = (1, 2, 3, 4, 5)
fruits = ("apple", "banana", "cherry")
```

tuple() 関数を使う

```python
# 文字列をタプルに変換
string_to_tuple = tuple("hello")
print(string_to_tuple) 

# リストをタプルに変換
list_to_tuple = tuple([1, 2, 3])
print(list_to_tuple) 

# range()関数で生成された範囲をタプルに変換
range_to_tuple = tuple(range(5))
print(range_to_tuple)  

# tuple()関数を使った空のタプル
empty_tuple = tuple()
print(empty_tuple) 
```

同じ値を繰り返して作成（初期化）する

```python
none_tuple = (None,) * 5
print(none_tuple)

zeros = (0,) * 10
print(zeros)

#このへんはリストと同じ。
```

他のシーケンスをタプルにコピーして作成

タプルは不可変（イミュータブル）なためコピー元から独立する。

```python
original_list = [1, 2, 3]
# listをタプルに変換
copied_tuple = tuple(original_list)
print(copied_tuple)  

original_tuple = (4, 5, 6)
copied_tuple2 = tuple(original_tuple)
print(copied_tuple2)
```

タプル　インデックスによるアクセス

```python
fruits = ("apple", "banana", "cherry")

# インデックス0（最初の要素）にアクセス
print(fruits[0])     #[]でインデックスを指定。

# インデックス-1（最後の要素）にアクセス
print(fruits[-1])  
```

タプル　スライス（部分的な要素の取得）

```python
numbers = (0, 1, 2, 3, 4, 5)

# インデックス1から3の要素を取得（3は含まない）
print(numbers[1:3]) 
print(numbers[::-1]) #逆順に
print(numbers)  #元の要素は変わらない
```

タプル　要素の検索

in キーワード で確認

index() メソッド を使用して特定の値が最初に現れる位置を取得

```python
fruits = ("apple", "banana", "cherry")
print("apple" in fruits)
print(fruits.index("banana"))  #index() メソッドを使って
```

タプル　要素のカウント

+ count() メソッド利用

```python
numbers = (1, 2, 3, 2, 4, 2, 5)

# 2がタプルに何回登場するか
print(numbers.count(2))  
```

タプルの結合

+ +演算子を使って結合。

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
tuple3 = tuple1 + tuple2
print(tuple3)
```
