一般的には with 構文を使用。close() 関数で閉じなくてもよくて楽

```python
with open(ファイルパス, mode=モード, encoding=エンコーディング) as file:
```

```python
with open("data/sample.txt", "r") as file:
# 配下のフォルダから書き始める。親フォルダは「..」で表す
```

モードの種類
"r"：読み込み専用。ファイルがないとエラー。
"w"：書き込み専用。既存ファイルを上書き。
"a"：追記専用。ファイルの末尾に追加される。
"r+"：読み書き可能。ファイルがないとエラー。

エンコーディング
encoding="utf-8"　を指定するのが一般的。


エラーハンドリング（例外処理）
ファイル操作は、try - except でエラーハンドリングを行うことが推奨。
ファイルが見つかりません、等。

```python
try:
    with open("non_existent_file.txt", "r") as file:
        data = file.read()
except FileNotFoundError:
    print("ファイルが見つかりません。")
```

#### Lesson 3　Chapter 8.2　テキストファイルの操作

+ テキストファイルへの書き込み

文字列の書き込み・・・write()メソッド。

```python
# 存在しないファイルを開けると、ファイルが新しく作成される。

# "w"モード。　全上書き。元の中身は消去される。
with open("sample.txt", "w") as file:
    file.write("Hello, World!")

# "a"モード。　追記モード。元の中身が残る。　最後に改行
with open("sample.txt", "a") as file:
    file.write("Hello, World!\n")

# 複数行を一度に書き込む場合は、writelines()メソッド。
lines = ["First line\n", "Second line\n", "Third line\n"]
with open("sample.txt", "a") as file:
    file.writelines(lines)
```

+ テキストファイルの読み込み

```python
# ファイルを読み込みモード "r" で★全部読み込む

# .read()メソッドで読み込んだものを変数へ渡す
# 注意：ファイルが非常に大きい場合はメモリを大量に消費する
with open("sample.txt", "r") as file:
    data = file.read()
    print(data)
```

```python
# ファイルを読み込みモード "r" で★一行ずつ読み込んでリスト化
# .readlines()メソッドで読み込んだものを1行ずつのリストとして変数へ渡す

with open("sample.txt", "r") as file:
    data = file.readlines()
    print(data)

    for index, line in enumerate(data):  #
        print(f"{index}:{line.strip()}") #.strip() で末尾の空白や改行が削除
```

+ 逐次行の読み込み

大きなファイルを扱う場合や、行ごとに処理を行いたい場合は、for ループを使って1行ずつ読み込む方法が便利です。

```python
# 読んで出力読んで出力を繰り返すからメモリを食いにくい？
with open ("sample.txt", "r") as file:
    for line in file:
        print(line.strip())
```

#### Lesson 3　Chapter 8.3　CSVファイルの操作

CSVファイルの構造と用途

```python
CSVファイルの中身の例：
name,age,city     #ヘッダー行
Alice,30,New York
Bob,25,Los Angeles
Charlie,35,Chicago
```

+ CSVファイルへのリストの書き込み

csv.writer（リスト形式）　

```python
#　リスト形式のデータをCSVファイルに書き込む

import csv
data = [
    ["name", "age", "city"],
    ["Alice", "30", "New York"],
    ["Bob", "25", "Los Angeles"],
    ["Charlie", "35", "Chicago"]
]

#newline=""とすることで、空行が挿入されるのを防ぐ
with open("sample1.csv", mode="w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file) #書き込み用のオブジェクトを作って変数に代入
    writer.writerows(data) 
```

+ 辞書をCSVファイルに書き込む

csv.DictWriter　（辞書形式）

```python
import csv

# 辞書のリスト
data = [
    {"name": "Alice", "age": "30", "city": "New York"},
    {"name": "Bob", "age": "25", "city": "Los Angeles"},
    {"name": "Charlie", "age": "35", "city": "Chicago"}
]

with open("sample2.csv", mode="w", encoding="utf-8", newline="")as file:
    fieldnames = ["name", "age", "city"]  #ヘッダ。辞書のキーと一致させる
    writer = csv.DictWriter(file, fieldnames = fieldnames) #辞書形式で
    writer.writeheader() #ヘッダー行の書き込み
    writer.writerows(data) #値を書き込み

# DictWriter は、辞書形式のデータをCSVに書くときに便利
# fieldnames で列の順番をコントロールできる
# writeheader() を忘れずに・・・
```

+ CSVファイルの読み込み

```python
import csv

with open("sample1.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)

# row[0] で名前、row[1] で年齢、row[2] で都市のように
# インデックスを指定して特定のデータの参照も可能
```

+ 辞書にCSVファイルを読み込む

```python
import csv

with open("sample2.csv", mode="r", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)

#　row["name"] で「name」列のデータが参照できる。
# 各データをわかりやすく扱える
```
