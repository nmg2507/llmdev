#### Lesson 3　Chapter 7.1　セットを作成する

セットとは：　順序なし＆重複なしの値の集合。

集合演算（和・積・差など）に強い。

+ 中括弧 {} で定義する

セットの変数名 = {要素1, 要素2, 要素3, ...}

```python
# 数値を格納したセット
numbers = {1, 2, 3, 4, 5}
print(numbers) 

# 異なるデータ型の値を格納したセット
mixed_set = {"apple", 42, (1, 2, 3)}
print(mixed_set) 

# 空のセット（空の辞書と区別するために set() を使用）
empty_set = set()
print(empty_set) 
```

+ set() 関数を使う

```python
# リストからセットを作成
numbers_list = [1, 2, 3, 2, 1]
unique_numbers = set(numbers_list)
print(unique_numbers)  # {1, 2, 3}

# 文字列からセットを作成
letters = set("hello")
print(letters)  # 順番もない。h,e,l,o　と表示される保証はない。
```

#### Lesson 3　Chapter 7.2　セットの基本操作

+ セットの要素の追加

```python
fruits = {"apple", "banana"}
fruits.add("cherry")
print(fruits) 

# すでに存在する要素を追加してもセットには影響なし
fruits.add("apple")
print(fruits)  
```

+ セットの要素の削除

```python
fruits = {"apple", "banana", "cherry"}

# 特定の要素を削除　remove
fruits.remove("banana")
print(fruits)  

# 存在しない要素の削除　　discard
fruits.discard("orange")  # エラーにならない
print(fruits)  

# 任意の要素を削除
removed_item = fruits.pop()
print(fruits)  # {'cherry'}（'apple' が削除されたと仮定）
print(removed_item)  

# セットを空にする
fruits.clear()
print(fruits) 
```

+ セットの集合演算（和・積・差・対称差）

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# 和集合　（union() または |）：2つのセットの要素すべて
print(a | b)  

# 積集合　（intersection() または &）：共通する要素のみ
print(a & b)  

# 差集合　（difference() または -）：片方にのみ存在する要素
print(a - b)  

# 対称差集合　（symmetric_difference() または ^）：一方にのみ存在する要素
print(a ^ b)  
```

+ セットの要素の存在確認　　in　キーワード

```python
fruits = {"apple", "banana", "cherry"}

# "apple" がセットにあるか確認
print("apple" in fruits) 

# "grape" がセットにあるか確認
print("grape" in fruits) 
```
