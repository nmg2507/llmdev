Lesson 3　Chapter 3
文字列でシーケンスの基本を学

```python
#インデックスによるアクセス

s = "abcdefghijklmnopqrstuvwxyz"
print(s)
print(s[3])
print(s[26]) #26番目は存在しない
```

```python
#部分的な要素の取り出し（スライス）
s = "abcdefghijklmnopqrstuvwxyz"
print(s)
print(s[3:5])
print(s[-5:-1])
print(s[15:])  #片方省略すると最後まで表示される
print(s[0:10:2]) #増分が指定できる（+２とび）
```

```python
#これを利用して回文を判定する　（全部ひっくりかえしても同じ）
s = "にんてんどううどんてんに"
if( s == s[::-1]):   #開始と終了を指定しないと全てが対象
    print (s ,"は回文です")
else:
    print (s,"は回文ではありません")
```

Lesson 3　Chapter 3.3　反復処理（ループ）

```python
#　基本形
s = "Banana"
for char in s:   #for文 sの要素分だけ繰り返してくれる
                 #in はメンバーシップ演算子（含まれるか？を返す）
    print(char)
```

```python
#例２　条件付きの反復処理 ・・母音が含まれるかどうか？
s = "Banana"
vowels = "aiueoAIUEO" # 抽出する母音を変数へ入れておく
for char in s:
    if char in vowels:   #代入された文字が、vowelsに含まれるか判定
        print(char)   #含まれたら(trueなら)出力
```
