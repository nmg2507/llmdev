#### Lesson 3 Chapter 6.1 辞書を作成

+ 中括弧 {} で定義する
+ dict() 関数を使う
+ dict.fromkeys() 関数を使う


 {} で辞書を定義する

-- 辞書の変数名 = { キー1: 値1, キー2: 値2, ... }

```python
numbers = {"one": 1, "two": 2, "three": 3}
print(numbers) 
print(type(numbers))

# 異なるデータ型の値を格納した辞書
person = {"name": "Alice", "age": 25, "is_student": False}
print(person)  # {'name': 'Alice', 'age': 25, 'is_student': False}

# 空の辞書
empty_dict = {}
print(empty_dict)  # {}
```

+ dict() 関数を使う

```python
# キーワードと引数で辞書作成 
person = dict(name="Bob", age=25, is_student=True)
print(person)

# タプルのリストから辞書を作成
list_tuple = [("apple", 100), ("banana", 80), ("orange",60)]
to_dict = dict(list_tuple)
print(to_dict)
```

+ dict.fromkeys() 関数を使う

補足：これはdictクラスそのもののﾒｿｯﾄﾞで、クラスメソッド。

```python
# キーを "key1" から "key4" として、全ての値を 0 に初期化
keys = ["k1", "k2", "k3", "k4"]
dict_keys = dict.fromkeys(keys,0)
print(dict_keys)
```

#### Lesson 3　Chapter 6.2　辞書の基本操作

+ キーによるアクセス

キーを指定することでその値を参照できる。
dictはミュータブル（可変）。直接書き換え可能。

```python
fruit_prices = {"apple": 100, "banana": 200, "cherry": 300}

# キー "apple" の値にアクセス
print(fruit_prices["apple"]) 

# キー "banana" の値を 250 に変更　書き換え
fruit_prices["banana"] = 250
print(fruit_prices)
```

+ 要素の追加・変更

```python
fruit_prices = {"apple": 100, "banana": 200}

#新しいキーを作成して追加。
fruit_prices["grape"] = 400  #新しいキーを作って、値を入れるだけ
print(fruit_prices)

# 既存のキー "banana" の値を変更
fruit_prices["banana"] = 250 #呼び出して新しい価格を入れるだけ
print(fruit_prices)
```

+ 要素の削除

```python
#　pop()メソッドで削除。
fruit_prices = {"peach": 100, "banana": 200, "apple": 400}
print(fruit_prices.pop("banana")) #バナナの要素を削除した
print(fruit_prices) #バナナのキーごと消えた


# 指定キーの要素を del キーワードで削除
del fruit_prices["peach"]
print(fruit_prices)  # {'apple': 100}

# 辞書を全部空にする
fruit_prices.clear()
print(fruit_prices)
```

★補足　特別な構文として使うキーワード　for　、del、in　などは関数でもなくメソッドでもなく。pythonの基本的な文の一部。

+ 辞書のキーと値の取得

```python
fruit_prices = {"apple": 100, "banana": 200, "cherry": 300}

# キーのリストを取得
keys = fruit_prices.keys()
print(keys) 

# 値のリストを取得
values = fruit_prices.values()
print(values)  # dict_values([100, 200, 300])

# キーと値のペアを取得
items = fruit_prices.items()
print(items)  
# dict_items([('apple', 100), ('banana', 200), ('cherry', 300)])
```

+ キーの存在確認

in キーワードを使用。

```python
fruit_prices = {"apple": 100, "banana": 200, "cherry": 300}

# "apple" が辞書にあるか確認
print("apple" in fruit_prices) 
```
