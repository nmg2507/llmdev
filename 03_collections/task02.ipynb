課題：単語の文字数をカウントしよう

```python
""" 仕様
英単語の入力を受け付けます。
未入力でリターン（エンター）キーを押されるまで、英単語の受付を
繰り返します。受付が完了したら、入力された英単語をABC順に
並べ替えて表示します。
また、すべての英単語に含まれるアルファベットごとの数も、値の多い順に表示します。
"""

#word = ""
alphabet = "abcdefghijklmnopqrstuvwxyz"
alphabet_list = list(alphabet)
alphabet_dict = dict.fromkeys(alphabet_list,0)
#print(alphabet_dict)

word_list = []
while True:
    word = input("英単語を入力してください：")
    if word == "":
        break
    else:
        word_list.append(word)
        for char in word.lower():
            if char in alphabet_dict:  #辞書に含まれない文字のエラー回避
                alphabet_dict[char] += 1

print("入力した英単語：",sorted(word_list))

#dictの値の大きい順で並び替えた表示用の辞書 2 を作る
#.itemsでペアにして、値の降順にならべる
alphabet_dict2 = sorted(alphabet_dict.items(), key=lambda x: x[1], reverse=True)


for key, value in alphabet_dict2:   #ソートされた辞書2の要素数分繰り返す
    if value > 0:  #値が０のものは表示させない。
        print (f"{key}が{value}個ありました") 
```
