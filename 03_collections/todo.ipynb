TODOリストアプリを作成しよう

```python
"""
TODOリストアプリの仕様
このTODOリスト管理アプリはノートブックで動作します。
機能は1～4の番号を入力して操作します。
1でTODO追加、2でTODO表示、4で終了・保存ができます。
入力したTODOはプログラム終了時にファイルへ保存されます。

予想
１、ファイルが無ければ空ファイルを作る(try-except)、あれば読み込む(open(・・・"r"))
２、ファイルがなかった場合は空のリストを作っておく（最後の書き込み時に書き込み）
３、ループを作って番号入力受付
４、力番号をif判別して各処理を実施
５、番号ごとの処理を作成

振り返り
１、空ファイルは最初に作らなくても最後の書き込み時に自動生成する
２、空リストは要る
３、readlines() でリストに再格納しても? ← こっちのほうがメモリ食う


"""
```

```python
# 試行錯誤。。。
try:
    with open("todo_list.txt", "r", encoding = "utf-8") as file:
        todo_list = [line.strip() for line in file]
except FileNotFoundError:
        todo_list = []

print("【Todoリスト】")

while True:
     print("メニュー番号を入力してください")
     print("[1:追加  2:表示  4:保存と終了]")
     num= input("メニュー番号")

     if num == "1":
          print("追加" ,flush=True)
          add_todo = input("追加するTodo：")
          print("追加するTODO:", add_todo, flush=True) #flush/リアルタイム性が必要な時
          todo_list.append(add_todo)
     elif num == "2":
          print("表示" ,flush=True)
          for index,line in enumerate(todo_list,start=1):
               print(index,line,"\n", flush=True)
     elif num == "4":
          print("保存と終了" ,flush=True)
          with open("todo_list.txt", "w", encoding="utf-8") as file:
               file.writelines(f"{line}\n" for line in todo_list)
               print("Todoリストを保存して終了します")
               break #ループ終了
     else:
          print("1,2,4の数字のいずれかを入力してください\n",flush=True)


```

##### （課題）削除機能を足す

```python
# 3:削除の追加版 
filename = "todo_list.txt"    # TODOリストを格納するリスト
try:
    # ファイルが存在したら内容を読込
    with open(filename, "r", encoding="utf-8") as file:
        todo_list = [line.strip() for line in file]  #リストに格納。
except FileNotFoundError:
    # ファイルが存在しない場合は新しいリストを作成
    todo_list = []

# メニュー表示
print("【TODOリスト】")

while True:
    print("番号を入力してください[1:追加、2:表示、 3:削除、 4:保存と終了  ]:", end="", flush=True)
    choice = input("")

    if choice == "1":
        print("1:追加", flush=True)
        add_todo = input("追加するTODO: ")
        print("追加するTODO:", add_todo, flush=True)
        todo_list.append(add_todo)

    elif choice == "2":
        print("2:表示", flush=True)
        for i, todo in enumerate(todo_list, start=1):
            print(f"{i}. {todo}")

    elif choice == "3":
        print("3:削除  （削除する行を下記から選んでください）\n")
        for i, todo in enumerate(todo_list, start=1):
            print(i,todo)
        try:
            del_i = int( input("削除する行番号を選んでください"))
            todo_list.pop(del_i - 1)
        except:
            print("リストに存在しない番号です")
            
        print(del_i,"番が削除されました\n")
        with open(filename,"w",encoding="utf-8") as file:
            file.writelines(f"{todo}\n" for todo in todo_list)

    elif choice == "4":
        print("4:保存と終了", flush=True)
        with open(filename, "w", encoding="utf-8") as file:
            file.writelines(f"{todo}\n" for todo in todo_list)
        print("TODOリストを保存して終了します。", flush=True)
        break

    else:
        print("1-4の数字を入力してください。", flush=True)
```
