TODOリストアプリを作成しよう

```python
"""
TODOリストアプリの仕様
このTODOリスト管理アプリはノートブックで動作します。
機能は1～4の番号を入力して操作します。
1でTODO追加、2でTODO表示、4で終了・保存ができます。
入力したTODOはプログラム終了時にファイルへ保存されます。

予想
１、ファイルが無ければ空ファイルを作る(try-except)、あれば読み込む(open(・・・"r"))
２、ファイルがなかった場合は空のリストを作っておく（最後の書き込み時に書き込み）
３、ループを作って番号入力受付
４、力番号をif判別して各処理を実施
５、番号ごとの処理を作成

振り返り
１、空ファイルは最初に作らなくても最後の書き込み時に自動生成する
２、空リストは要る
３、readlines() でリストに再格納しても? ← こっちのほうがメモリ食う


"""
```

```python
# 試行錯誤。。。
try:
    with open("todo_list.txt", "r", encoding = "utf-8") as file:
        todo_list = [line.strip() for line in file]
except FileNotFoundError:
        todo_list = []

print("【Todoリスト】")

while True:
     print("メニュー番号を入力してください")
     print("[1:追加  2:表示  4:保存と終了]")
     num= input("メニュー番号")

     if num == "1":
          print("追加" ,flush=True)
          add_todo = input("追加するTodo：")
          print("追加するTODO:", add_todo, flush=True) #flush/リアルタイム性が必要な時
          todo_list.append(add_todo)
     elif num == "2":
          print("表示" ,flush=True)
          for index,line in enumerate(todo_list,start=1):
               print(index,line,"\n", flush=True)
     elif num == "4":
          print("保存と終了" ,flush=True)
          with open("todo_list.txt", "w", encoding="utf-8") as file:
               file.writelines(f"{line}\n" for line in todo_list)
               print("Todoリストを保存して終了します")
               break #ループ終了
     else:
          print("1,2,4の数字のいずれかを入力してください\n",flush=True)


```
