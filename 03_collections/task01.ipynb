学生の成績分析ツールを作成しよう

```python
grades = [85, 90, 78, 92, 88] #学生の成績

average = sum(grades)/ len(grades) 
max_grades = max(grades)
min_grades = min(grades)

count = 0  #成績が平均以上の学生数をカウント
for i in grades:
    if i >= average:
        count += 1

#以下出力
print("平均点：", average)
print("最高点：", max_grades)
print("最低点：", min_grades)
print("平均以上の学生数：", count)
```
