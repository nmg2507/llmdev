「回文を判定する処理」

```python
def is_palindrome(str):  #関数名は、動作がわかりやすい名前に！
    if str == str[::1]:
        print(f"「{str}」は回文です")
    else:
        print(f"「{str}」は回文ではありません")

# 試してみる　（defで定義した内容が()の変数に対する関数として使える）
is_palindrome("にんてんどううどんてんに")
```

```python
# テキスト
def is_palindrome(s):
    if s == s[::-1]:
        print(f"「{s}」は回文です！")
    else:
        print(f"「{s}」は回文ではないです。")

is_palindrome("にんてんどううどんてんに")
is_palindrome("Pythonは便利だが遅い")
is_palindrome("no devil lived on")
```

#### Lesson 4 Chapter 2.3　引数（パラメータ）

```python
def greet(name):
    print(f"こんにちは、{name}さん！")

greet("山田")
```

```python
# 複数の引数も可能
def introduce(name, age, adr):
    print(f"こんにちは、{name}さん！ 年齢は{age}歳、{adr}にお住まいですね。")

introduce("yamada", "23", "足立区")
```

```python
# デフォルトの値もdef内に記述できる
def introduce(name="だれか", age="不明", adr="どこか"):
    print(f"こんにちは、{name}さん！ 年齢は{age}歳、{adr}にお住まいですね。")

introduce("yamada", "23", "足立区")
introduce()
```

+ 可変長引数（*args と **kwargs）

```python
# *args　複数の位置引数をタプルとして受け取り。
# ★関数がうけとるタプル内の順番は大事
# 「arg」は引数の意味

def add(*args):
    result = sum(args)
    print(f"合計:{result}")

add(1, 2, 3)    #リストではなくタプル（イミュータブルな引数）
add(5, 10, 15, 20)
```

```python
# **kwargs　複数のキーワード引数を、辞書として受け取る
# ”キーワード(タイトル)付き”　の辞書で渡すから、柔軟性が高い。
def introduce(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}") 

introduce(name="太郎", age=20, hobby="読書")
```

```python
#応用：辞書に「**」をつけて、dictの代数のまま渡すこともできる　
def introduce2(**kwargs):
    name = kwargs.get("name", "だれか") 
    age = kwargs.get("age", "不明")
    hobby = kwargs.get("hobby", "何か")
    print(f"{name}さんの年齢は{age}歳で、趣味は{hobby}。")

person = dict(name = "田中", age=26, hobby="サッカー")
introduce2(**person)
```

#### Lesson 4　Chapter 2.4　戻り値（return文）
関数の処理が終了するときに結果を返す　（変数で受け取れる）

```python
def add(a, b):
    result = a + b
    return result

total = add(3, 5) #returnで返された値を変数でうけとる
print(total)
```

returnの複数の戻り値

```python
# カンマ,で区切られた複数の戻り値を返す
def calculate(a, b):
    sum_result = a + b
    product_result = a * b
    return sum_result, product_result

#　同じ形(カンマで区切られた複数)の変数で戻り値をキャッチする
i, j = calculate(5, 6)
print(i, j)
```

None を返す関数

```python
# return 文がない関数は、デフォルトで None を返してる
def greet():
    print("Hello!")

result = greet()
print(result)
```

計算関数の結果を使って、別の計算を行う例：

```python
# 掛け算を行う関数
def multiply(a, b):
    return a * b

# 2乗を行う関数
def square(number):
    return multiply(number, number)

print(square(5))
```

（課題）ユーザー名のバリデーション（検証）

不正なユーザー名が入力されたらエラーを表示する、ユーザー名バリデーションの関数を作成してみましょう。

```python
"""
関数の仕様：
文字列を受け取り、バリデーションの結果をブール値で返す関数を作成してください。
バリデーションのルールは以下です。
・ユーザー名は5文字以上15文字以内。条件を満たさない場合は 
エラー: ユーザー名は5文字以上15文字以内で入力してください と表示して 
False を返してください。
・ユーザー名に空白は使えない。含まれている場合は エラー: ユーザー名に
空白を含めることはできません と表示して False を返してください。
・ユーザー名は英数字（アルファベットと数字）のみ使用可能。それ以外の
文字が含まれている場合は エラー: ユーザー名は英数字のみ使用できます と
表示して False を返してください。
すべての条件を満たす場合は、Trueを返す（※成功メッセージは表示しない）
"""

# 関数１：5文字以上15文字以内ならTrue、違えばFalse
def right_length(username):
    if 5 <=len(username) <=15:
        return True
    else:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False

# 関数２：空白文字を含めばFalse,そうでなければTrue
def no_space(username):
    if " "in username or "　" in username:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    else:
        return True

# 関数３：英数字のみならTrue,そうでなければFalse
def is_alnum(username):
    if username.isalnum() == True:  #英数字のみならTrueを返すstrのメソッド
        return True
    else:
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False

#関数４：関数１～３がTrue　ならTrue
def check_all(username):
    if right_length(username) and no_space(username) and is_alnum(username):
        return True
    else:
        return False

# 変数 username へ文字を格納する。
try:
    username = input("ユーザー名を入力してください")
    is_ok = check_all(username)
    if is_ok ==True:
        print("ユーザー名は有効です")
    else:
        print("不正なユーザー名です")
except:
    print("想定外のエラーのようです。もう一度やりなおしてください")
```

#### Lesson 4　Chapter 3 モジュールとライブラリの活用

+ 基本　import モジュール名

```python
import math

# mathモジュールの平方根メソッド
print(math.sqrt(16))  
```

+ エイリアスを使う

import　モジュール名　as　別名


```python
import datetime as dt

# 現在の日時を取得
now = dt.datetime.now()
print("現在の日時:", now)
```

+ 特定の関数や変数だけをインポート

from モジュール名 import 関数名

```python
from math import sqrt

# 直接指定の関数だけを呼び出せる。メモリの節約にもなる。
# また、　math.sqrt　ではなく、★直接関数として　sqrt()　と書ける
# デメリットは、どこのモジュールの関数や変数かが、わかりにくくなる
print(sqrt(16))  
```

```python
from math import sqrt, pi    # 2ついっぺんにインポートもできる

# sqrtとpiを直接呼び出し
print(sqrt(16)) 
print(pi)       
```
