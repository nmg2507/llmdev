課題：銀行口座の管理

ユーザーが残高を直接変更できないように、カプセル化を使って残高を管理。

預け入れ（deposit）や引き出し（withdraw）が正常に行えるよう

残高不足の場合はエラーを表示する機能も実装

```python
 # BankAccount クラスの定義
class BankAccount:
    #残高：balanceはカプセル化で保護
    def __init__(self, balance):  #初期残高値を _balanceからうけとる
    # ↓属性名balanceの前に '_' をつける=外部からの直接アクセスを避ける意図を示す
        self._balance = balance

    # ゲッター　get_balance()メソッドでbalanceを参照
    def get_balance(self):
        return self._balance

    # セッター　deposit()とwithdraw()メソッドでアクセス可能に
    def deposit(self, amount):  # amount分の金額を預入
        if amount <= 0:  # 0より小さい場合はエラー(バリデーション)
            print("エラー: 正しい金額を入力してください")
            return
        self._balance += amount
        print(f"（{amount}円を預け入れ）") #任意

    def withdraw(self, amount): # amount分の金額を引出し
        if amount <= 0:  # 0以下のエラー
            print("エラー: 残高不足または無効な金額です")
            return

        if amount > self._balance:  # 残高不足のエラー
            print("エラー: 残高不足または無効な金額です")
            return

        self._balance -= amount  # 引き出し可能なら残高を更新
        print(f"（{amount}円を引き出し）")  #任意



# 呼び出し側 

# 初期残高1000(円)を引数に BankAccount クラスのインスタンス作成
account = BankAccount(1000)

# 500円を預け入れ
account.deposit(500)

# 200円を引き出し
account.withdraw(200)

# 現在の残高を表示（1300 と表示される）
print(account.get_balance())

# 1500円を引き出そうとする（エラー: 残高不足または無効な金額です。）
account.withdraw(1500)

# 預け入れ時に無効な金額（-500など）を指定した場合
account.deposit(-500)
```
