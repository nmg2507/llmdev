å¤‰ãªãªå€¤ãŒå…¥ã‚‰ãªã„ã‚ˆã†ã«å…ˆã«_nameã€€ã®ä»£æ•°ã§å—ã‘å–ã£ã¦ã€æ¡ä»¶ã§ãƒ•ã‚£ãƒ«ã‚¿ã™ã‚‹ãƒ‡ãƒ¼ã‚¿ã®æ­£ã—ã•ã‚’å®ˆã‚‹ï¼ˆï¼5æ­³ã€ã¨ã‹ã„ã†æ•°å­—ã‚’ã¯ã˜ã„ãŸã‚Šï¼‰

ä¾‹ãˆã°ã‚ã¨ã§çŠ¬ã®å¹´é½¢ã‚’äººé–“æ›ç®—ã—ãŸã„ã€ã¨ãªã£ã¦ã‚‚ã€ã‚²ãƒƒã‚¿ãƒ¼ã®ã¨ã“ã‚ã§èª¬æ˜ã‚’ãƒ—ãƒªãƒ³ãƒˆè¿½åŠ ã§ãã‚‹


@property ã‚’ä½¿ã£ã¦ã‚¹ãƒãƒ¼ãƒˆã«ä½¿ã†

ã‚²ãƒƒã‚¿ãƒ¼ã¨ã‚»ãƒƒã‚¿ãƒ¼ã‚’å®šç¾©ã—ã¤ã¤ã€ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã®å±æ€§ã«
ç›´æ¥ã‚¢ã‚¯ã‚»ã‚¹ã™ã‚‹ã‚ˆã†ã«è¦‹ã›ã‹ã‘ã‚‰ã‚Œã‚‹ï¼ˆè£ã§ã®æŒ™å‹•ã¯ä¸€ç·’ï¼ï¼‰

```python
class Dog:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    # ãƒ—ãƒ­ãƒ‘ãƒ†ã‚£ã‚’ä½¿ã£ãŸã‚²ãƒƒã‚¿ãƒ¼ã€€*getterã€€ã¨ã¯æ›¸ã‹ãªã„
    @property
    def name(self):
        return self._name

    # ãƒ—ãƒ­ãƒ‘ãƒ†ã‚£ã‚’ä½¿ã£ãŸã‚»ãƒƒã‚¿ãƒ¼
    @name.setter
    def name(self, name):
        if len(name) > 0:
            self._name = name
        else:
            print("åå‰ãŒç©ºã§ã™ã€‚æ­£ã—ã„åå‰ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„ã€‚")

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, age):
        if age > 0:
            self._age = age
        else:
            print("å¹´é½¢ã¯1ä»¥ä¸Šã«è¨­å®šã—ã¦ãã ã•ã„ã€‚")

dog = Dog("ãƒãƒ", 5)
print(dog.name)  # ãƒãƒã¨è¡¨ç¤ºã•ã‚Œã‚‹ï¼ˆã‚²ãƒƒã‚¿ãƒ¼ï¼‰
dog.name = "ã‚¸ãƒ§ãƒ³"  # åå‰ã‚’ã€Œã‚¸ãƒ§ãƒ³ã€ã«å¤‰æ›´ï¼ˆã‚»ãƒƒã‚¿ãƒ¼ï¼‰
print(dog.name)  # ã‚¸ãƒ§ãƒ³ã¨è¡¨ç¤ºã•ã‚Œã‚‹
```

#### Lesson 5ã€€Chapter 5.3ã€€super() ãƒ¡ã‚½ãƒƒãƒ‰
å­ã‚¯ãƒ©ã‚¹ã‹ã‚‰è¦ªã‚¯ãƒ©ã‚¹ã®ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å‘¼ã³å‡ºã—ãŸã„å ´åˆã¯ã€super()ã€€ã§

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        return "ä½•ã‹éŸ³ã‚’å‡ºã—ã¦ã„ã¾ã™ã€‚"

class Dog(Animal):
    def speak(self):
        parent_message = super().speak()  # è¦ªã‚¯ãƒ©ã‚¹ã® speak ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å‘¼ã³å‡ºã™
        return f"{parent_message} ãã—ã¦ã€{self.name} ã¯ãƒ¯ãƒ³ãƒ¯ãƒ³ã¨å ãˆã¦ã„ã¾ã™ï¼"

dog = Dog("ãƒãƒ", 3)
print(dog.speak())  # ä½•ã‹éŸ³ã‚’å‡ºã—ã¦ã„ã¾ã™ã€‚ãã—ã¦ã€ãƒãƒ ã¯ãƒ¯ãƒ³ãƒ¯ãƒ³ã¨å ãˆã¦ã„ã¾ã™ï¼
```

#### Lesson 5ã€€Chapter 6ã€€å¤šæ…‹æ€§ï¼ˆãƒãƒªãƒ¢ãƒ¼ãƒ•ã‚£ã‚ºãƒ ï¼‰

```python
# è¦ªã‚¯ãƒ©ã‚¹ã«å¯¾ã—ã¦å­ã‚¯ãƒ©ã‚¹ã‚’è¤‡æ•°ã¤ãã£ã¦ã€
# åŒã˜ãƒ¡ã‚½ãƒƒãƒ‰ã§ã‚¯ãƒ©ã‚¹ã”ã¨ã«é•ã†å‹•ä½œã‚’ã•ã›ã‚‹ã€‚
# ç•°ãªã‚‹ã‚¯ãƒ©ã‚¹ã‚’ä¸€æ‹¬ã—ã¦å‡¦ç†ã™ã‚‹éš›ã«ä¾¿åˆ©ã€‚

class Animal:
    def speak(self):
        return "ä½•ã‹éŸ³ã‚’å‡ºã—ã¦ã„ã¾ã™ã€‚"

class Dog(Animal):
    def speak(self):
        return "ãƒ¯ãƒ³ãƒ¯ãƒ³ï¼"

class Cat(Animal):
    def speak(self):
        return "ãƒ‹ãƒ£ãƒ¼ï¼"

animals = [Dog(), Cat()] #ãƒªã‚¹ãƒˆå½¢å¼ã§ã€ãã‚Œãã‚Œã®ã‚¯ãƒ©ã‚¹ã®ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã‚’ç”Ÿæˆ
#ï¼ˆå®šç¾©ã•ã‚Œã¦ã„ãªã„ã‹ã‚‰åå‰ã¯ç„¡ã„ãŒã€çŠ¬1åŒ¹ã€çŒ«1åŒ¹ã®ã‚¤ãƒ¡ãƒ¼ã‚¸ï¼‰

for animal in animals:
    print(animal.speak())  # å„ã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆã«å¿œã˜ãŸãƒ¡ã‚½ãƒƒãƒ‰ãŒå‘¼ã°ã‚Œã‚‹
```

```python
#ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ãƒ¡ã‚½ãƒƒãƒ‰ã€€ã€€ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã”ã¨ã«å‹•ä½œã™ã‚‹
class Dog:
    def __init__(self, name):
        self.name = name  # ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã€Œå±æ€§ã€

    def bark(self):  # ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ãƒ¡ã‚½ãƒƒãƒ‰
        return f"{self.name} ãŒãƒ¯ãƒ³ãƒ¯ãƒ³ã¨å ãˆã¾ã™ï¼"

dog = Dog("ãƒãƒ")
print(dog.bark())  # ãƒãƒ ãŒãƒ¯ãƒ³ãƒ¯ãƒ³ã¨å ãˆã¾ã™ï¼
```

```python
#é™çš„ãƒ¡ã‚½ãƒƒãƒ‰
# ç‰¹å®šã®ã‚¯ãƒ©ã‚¹ã«é–¢é€£ã™ã‚‹ãŒã€ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã‚„ã‚¯ãƒ©ã‚¹è‡ªä½“ã«ã¯
# ä¾å­˜ã—ãªã„å‡¦ç†ã‚’è¡Œã†ãŸã‚ã®ãƒ¡ã‚½ãƒƒãƒ‰ã€‚
# ç¬¬ä¸€å¼•æ•°ã¨ã—ã¦ self ã‚„ cls ã‚’å–ã‚‰ãªã„ãŸã‚ã€ä»–ã®å±æ€§ã«ã‚¢ã‚¯ã‚»ã‚¹ã§ãã¾ã›ã‚“ã€‚
# é™çš„ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å®šç¾©ã™ã‚‹ã«ã¯ã€# @staticmethod ãƒ‡ã‚³ãƒ¬ãƒ¼ã‚¿ã‚’ä½¿ç”¨

class Dog:
    @staticmethod
    def info():
        return "çŠ¬ã¯äººé¡ã®æœ€è‰¯ã®å‹ã§ã™ã€‚"

print(Dog.info())  # çŠ¬ã¯äººé¡ã®æœ€è‰¯ã®å‹ã§ã™ã€‚
```

__str__: æ–‡å­—åˆ—è¡¨ç¾ã€‚print() ã‚„ str() ã‚’ä½¿ã£ãŸã¨ãã«å‘¼ã³å‡ºã•ã‚Œã¾ã™ã€‚

__repr__: é–‹ç™ºè€…å‘ã‘ã®æ–‡å­—åˆ—è¡¨ç¾ã€‚repr() ã‚’ä½¿ã£ãŸã¨ãã‚„ã€ãƒ‡ãƒãƒƒã‚°ã®éš›ã«å½¹ç«‹ã¡ã¾ã™ã€‚

__len__: é•·ã•ã€‚len() é–¢æ•°ã‚’ä½¿ã£ãŸã¨ãã«å‘¼ã³å‡ºã•ã‚Œã¾ã™ã€‚

__eq__: ç­‰ä¾¡æ¯”è¼ƒã€‚== æ¼”ç®—å­ã‚’ä½¿ã£ãŸã¨ãã«å‘¼ã³å‡ºã•ã‚Œã¾ã™ã€‚

```python
# ãƒã‚¸ãƒƒã‚¯ãƒ¡ã‚½ãƒƒãƒ‰ã¯ã€ç‰¹å®šã®ã€ŒPythonã®çµ„ã¿è¾¼ã¿æ©Ÿèƒ½ã€ãŒå®Ÿè¡Œã•ã‚ŒãŸã¨ãã«ã€
# è‡ªå‹•çš„ã«å‘¼ã³å‡ºã•ã‚Œã‚‹**ã€Œè£å´ã®æ©Ÿèƒ½ã€**
# ã“ã¾ã”ã¾æŒ‡å®šã—ãªãã¦ã‚‚ä½¿ãˆã‚‹ã€‚


class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):  # print()ã‚„str()ã§è¡¨ç¤ºã•ã‚Œã‚‹æ–‡å­—åˆ—
        return f"çŠ¬ã®åå‰: {self.name}, å¹´é½¢: {self.age}æ­³"

    def __repr__(self):  # repr()ã‚„ãƒ‡ãƒãƒƒã‚°ã®éš›ã«è¡¨ç¤ºã•ã‚Œã‚‹æ–‡å­—åˆ—
        return f"Dog(name='{self.name}', age={self.age})"

    def __eq__(self, other):  # == æ¼”ç®—å­ã®æŒ™å‹•ã‚’å®šç¾©
        return self.age == other.age   #ã€Œå¹´é½¢ã ã‘ã‚’æ¯”è¼ƒã•ã›ã¦ã‚‹ã€

dog1 = Dog("ãƒãƒ", 5)
dog2 = Dog("ã‚¸ãƒ§ãƒ³", 5)

print(dog1)  # çŠ¬ã®åå‰: ãƒãƒ, å¹´é½¢: 5æ­³
print(repr(dog1))  # Dog(name='ãƒãƒ', age=5)
print(dog1 == dog2)  # Trueï¼ˆå¹´é½¢ãŒåŒã˜ãŸã‚ï¼‰
```

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "ãƒ¯ãƒ³ãƒ¯ãƒ³ï¼"

class Cat(Animal):
    def speak(self):
        return "ãƒ‹ãƒ£ãƒ¼ï¼"

def make_animal_speak(animal: Animal):
    print(animal.speak())

make_animal_speak(Dog())  # ãƒ¯ãƒ³ãƒ¯ãƒ³ï¼
make_animal_speak(Cat())  # ãƒ‹ãƒ£ãƒ¼ï¼
```
