インターネット通信とWebアプリの仕組み

ブラウザで https://www.yahoo.co.jp というURLにアクセスするとします。このとき、Webページが表示されるまでの過程を時系列に説明してください。ただし、下記に指定した用語をすべて使用してください。

ブラウザ
DNS
IPアドレス
Webサーバ
HTTPリクエスト
HTTPレスポンス

ブラウザで指定のWebページにアクセスするとき、ユーザーは”https://www.yahoo.co.jp” などのURLを指定します。URLはコンピュータがアクセスする宛先として解読できるIPアドレス（例: 182.22.24.124など）と対応しており、DNSサーバでIPアドレスへ変換されます。
ブラウザからのアクセスは、https://www.yahoo.co.jp のページがあるWebサーバへHTTPリクエストというメッセージを送信する形で行われます。それに対しWebサーバーからはリクエストに対応したHTTPレスポンスを返します。HTTPレスポンスはブラウザで解読されてWebページとして表示されます。
