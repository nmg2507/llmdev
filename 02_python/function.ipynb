#### Lesson 2 Chapter 8 組み込み関数

```python
# print関数　
print("Hello, Python!")
print(123)              
print("test", end="", flush=True) # end="" で改行をとめる。
print("test", flush=True) # flushをTrueにするとリアルタイムで表示されていく
# Pythonでは、print() の出力はすぐに画面に表示されるとは限らない.
# 「バッファリング」っていう仕組みがあり、、出力が一時的にため込まれてから
# まとめて表示されることがある。それをせず、リアルタイムに表示するということ。
# ループさせて使うときなどにも。
```

```python
#len()　データの長さ（要素数）を返す。文字列、リスト、タプル、辞書
print(len("Python"))     # 出力: 6（文字数）
print(len([1, 2, 3, 4])) # 出力: 4（リストの要素数）
```

```python
# type()　データの型を確認
print(type(123))  
print(type("Python"))  
print(type([1, 2, 3])) 
print(type({1, 2, 3}))
print(type({1: "a", 2: "b", 3: "c"}))
```

```python
# max() と min()
# 数値や文字列などの要素から、最大値や最小値を取得
print(max(1,4,8))
print(min(4,1,8))
print(max("c","a","b")) # 文字列形式ではアルファベット順に
print(min(["c","a","b"])) # リストでも同様
print(min([4,2,3,1,5]))
```

```python
#　range()　指定した範囲の整数の連続を生成。ループの反復処理などで利用
for i in range(4):         # 出力: 0 1 2 3 4
   print(i)

print(list(range(1, 10, 2))) # 出力: [1, 3, 5, 7, 9]
```

補足：range()　はイテラブル（順番に取り出せるものにする）な組み込みオブジェクト。

```python
#組み込み関数の練習問題　文字数をカウント

name = input("名前を入力してください: ")
print("文字数は：",len(name))
```

```python
#組み込み関数の練習問題　最大値と合計

nums = [10,20,30,40,50]
print("最大数は：",max(nums))
print("合計値は：",sum(nums))
```

```python
#組み込み関数の練習問題　そのほか
print (sorted([3,5,1,8]))   #sorted() 並び替え。
```

```python
#absは絶対値を返す
print(abs(-3))
# print (abs([-1,4,-6,2,-9]))  はエラー。　扱えるのは数字一つのみ。
# 複数の数値なら、下記のような方法で:
nums = [-1, 4, -6, 2, -9]
abs_values = [abs(n) for n in nums]   #内包表記
print(abs_values)  

# enumerate() インデックス番号をふる
fruits = ["りんご", "みかん", "バナナ"]
for i, one_fruit in enumerate(fruits,start=1):  
    #startは、番号を1から開始させる。無指定の場合、0からはじまる
    print(i,"：",one_fruit)
```
